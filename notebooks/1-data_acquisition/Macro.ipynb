```python
#this function retrieves data from FRED api and renames the column of interest
def fredata(series_name, variable, start = '2004-01-01', end = '2024-01-01'):
    data = fred.get_series(series_name, start, end)
    df = pd.DataFrame({variable: data})
    return df.rename_axis('Date')
```

## IR UK

Manually downloaded from Bank of England: https://www.bankofengland.co.uk/monetary-policy/the-interest-rate-bank-rate

```python
ir_uk = pd.read_csv('../../data/preparation/ir_uk.csv', delimiter=';')
```

```python
ir_uk.drop(['Unnamed: 2'], axis=1, inplace=True)

# Convert the 'Date' column to datetime
ir_uk['Date'] = pd.to_datetime(ir_uk['Date'], format='%d/%m/%Y')

# Set 'Date' as the index
ir_uk.set_index('Date', inplace=True)

# Create a date range that spans the full range of dates in the DataFrame
full_date_range = pd.date_range(start=ir_uk.index.min(), end='01/01/2024', freq='D')

# Reindex the DataFrame to the full date range
ir_uk = ir_uk.reindex(full_date_range)

# Forward fill the missing dates
ir_uk.ffill(inplace=True)

# Reset the index and convert the 'Date' column back to the desired format
ir_uk.reset_index(inplace=True)
ir_uk.rename(columns={'index': 'Date'}, inplace=True)
ir_uk['Date'] = ir_uk['Date'].dt.strftime('%Y/%m/%d')

ir_uk.set_index('Date', inplace=True)
ir_uk.rename(columns={'Official Bank Rate' : 'ir_uk'}, inplace=True)

ir_uk = ir_uk.query("'2004-01-01' <= index <= '2024-01-01'")
```
