```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
```

```python
df = pd.read_csv('../../data/df_joint.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
null_values = df.isnull().sum()
null_values
```

```python
# FunciÃ³n auxiliar para formatear el eje x
def format_xaxis(ax):
    ax.xaxis.set_major_formatter(DateFormatter("%Y"))
    ax.figure.autofmt_xdate()

# Celda 8: Visualizar las series temporales de los tipos de cambio
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['eurusd'], label='EUR/USD')
plt.plot(df['Date'], df['jpyusd'], label='JPY/USD')
plt.plot(df['Date'], df['gbpusd'], label='GBP/USD')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Exchange Rates Over Time')
plt.legend()
format_xaxis(plt.gca())
plt.show()
```

```python
# Celda 9: Visualizar las series temporales de los Ã­ndices bursÃ¡tiles
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['nq'], label='Nasdaq 100')
plt.plot(df['Date'], df['sp500'], label='S&P 500')
plt.plot(df['Date'], df['nikkei'], label='Nikkei 225')
plt.plot(df['Date'], df['ftse'], label='FTSE 100')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Stock Indices Over Time')
plt.legend()
format_xaxis(plt.gca())
plt.show()
```

```python
# Celda 10: Visualizar series temporales de PIB de diferentes regiones
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['gdp_us'], label='GDP US')
plt.plot(df['Date'], df['gdp_uk'], label='GDP UK')
plt.plot(df['Date'], df['gdp_jp'], label='GDP JP')
plt.plot(df['Date'], df['gdp_ez'], label='GDP EZ')
plt.xlabel('Date')
plt.ylabel('GDP')
plt.title('GDP Over Time')
plt.legend()
format_xaxis(plt.gca())
plt.show()
```
