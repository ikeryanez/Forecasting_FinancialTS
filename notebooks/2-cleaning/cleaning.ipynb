```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/df_joint.csv')
df.head()
```

```python
df.info()
```

```python
resize = ['gdp_us', 'gdp_jp', 'm0_us', 'm0_jp']
df[resize].head()
```

```python
for var in resize:
    df[var] = df[var].apply(lambda x: x * 1000)
df[resize].head()
```

```python
df.drop(columns=['ir_deposit_ez', 'ir_lending_ez'], inplace=True)
```

```python
df.info()
```

```python
df.rename(columns={'ir_MRO_ez': 'ir_ez'}, inplace=True)
```

```python
df.rename(columns={'Date': 'date'}, inplace=True)
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')
```

```python
df['m0_jp'] = df['m0_jp'].astype('float64')
df['ir_uk'] = df['ir_uk'].str.replace(',', '.').astype('float64')
```

```python
df.set_index('date', inplace=True)
```

```python
df.head()
```

```python
df.to_csv('../../data/df_clean.csv', index=True, index_label='date')
```
