```python
import yfinance as yf
```

# USD vs EUR

```python
usdeur = yf.Ticker("EUR=X")
usdeur_10y = usdeur.history(period="10y") #Return a pandas dataframe of daily info
display(usdeur_10y)
```

```python
usdeur_10y.reset_index(inplace=True)
usdeur_10y.head()
```

```python
usdeur_10y.drop(columns=['Volume', 'Dividends', 'Stock Splits'])
usdeur_10y.rename(columns={'Date':'date', 'Open':'open', 'High':'high', 'Low':'low', 'Close':'close'}, inplace=True)
```

```python
usdeur_10y.head()
```

```python
usdeur_10y.to_csv('data/usdeur.csv', index=False)
```
